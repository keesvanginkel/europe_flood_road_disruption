### Load descriptive stats

```python
import json
from pathlib import Path
import pandas as pd
dr = Path('version3')
```

```python
contents = {}
for file in dr.iterdir():
    if file.suffix == '.json':
        contents[file.stem] = file
contents
```

```python
### Initiate dataframe from first json
first = list(contents.keys())[0]
with open(contents[first]) as file:
    file_content = json.load(file)
file_content
results = pd.DataFrame(file_content,index=[first])

### Iterate over the rest of the information

for country, path in contents.items():
    if country == first: #skip the json from which the DF was configurated
        continue
    else:
        #print(country,path)
        with open(contents[country]) as file:
            file_content = json.load(file)
        results = results.append(pd.DataFrame(file_content,index=[country]))
results
```

```python
results.columns
```

```python
r = results.sort_values(by='active_country_area_km2')
r['i_hit_PRIM_LEN%'] = r['i_hit_PRIM_LEN'] / r['Len_PRIM'] * 100
r['i_hit_PRIM_COUNT%'] = r['i_hit_PRIM_COUNT'] / r['Edgecount_PRIM'] * 100
r['i_0cm_PRIM_LEN%'] = r['i_0cm_PRIM_LEN'] / r['Len_PRIM'] * 100
r['i_0cm_PRIM_COUNT%'] = r['i_0cm_PRIM_COUNT'] / r['Edgecount_PRIM'] * 100

r['i_hit_ALL_LEN%'] = r['i_hit_ALL_LEN'] / r['Len_tot__km'] * 100
r['i_hit_ALL_COUNT%'] = r['i_hit_ALL_COUNT'] / r['Edge_count'] * 100
r['i_0cm_ALL_LEN%'] = r['i_0cm_ALL_LEN'] / r['Len_tot__km'] * 100
r['i_0cm_ALL_COUNT%'] = r['i_0cm_ALL_COUNT'] / r['Edge_count'] * 100


r[['i_hit_PRIM_LEN%','i_hit_PRIM_COUNT%','i_0cm_PRIM_LEN%','i_0cm_PRIM_COUNT%',
   'i_hit_ALL_LEN%' ,'i_hit_ALL_COUNT%' ,'i_0cm_ALL_LEN%' ,'i_0cm_ALL_COUNT%']] = \
r[['i_hit_PRIM_LEN%','i_hit_PRIM_COUNT%','i_0cm_PRIM_LEN%','i_0cm_PRIM_COUNT%',
   'i_hit_ALL_LEN%' ,'i_hit_ALL_COUNT%' ,'i_0cm_ALL_LEN%' ,'i_0cm_ALL_COUNT%']].apply(lambda x: round(x,1))
#r[['country_name','Len_PRIM','i_hit_PRIM_LEN%','i_hit_PRIM_COUNT%','i_0cm_PRIM_LEN%','i_0cm_PRIM_COUNT%']]
r[['country_name','Len_tot__km','i_hit_ALL_LEN','i_hit_ALL_LEN%','i_hit_ALL_COUNT%','i_0cm_ALL_LEN%','i_0cm_ALL_COUNT%']]
```
